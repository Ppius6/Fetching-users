```python
import requests
import json

# Set the API endpoint URL and parameters
url = 'https://randomuser.me/api'
params = {
    'results': 100,  # number of results to return
    'gender': 'male'  # filter by gender (male)
}

# Make the API request and get the response
response = requests.get(url, params=params)

# Parse the response JSON data
data = json.loads(response.text)

# Extract the user information from the JSON data
users = data['results']

# Loop through the users and print their information
for i, user in enumerate(users):
    name = f"{user['name']['title']} {user['name']['first']} {user['name']['last']}"
    email = user['email']
    phone = user['phone']
    print(f"{i+1}. {name}, email: {email}, phone: {phone}")
```

```python
# Importing pandas
import pandas as pd

# Creating a DataFrame from the users list
pd.DataFrame(users)
```
